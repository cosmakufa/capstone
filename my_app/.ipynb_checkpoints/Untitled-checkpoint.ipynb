```python
def new_x(x):
    nut = x['clean_json']
    for obj in nut:
        title = obj['title']
        amount = obj['amount']
        percentOfDailyNeeds = obj['percentOfDailyNeeds']
        unit = obj['unit']
        x[title+'_amount'] = (amount)
        x[title+'_percentOfDailyNeeds'] = (percentOfDailyNeeds)
        x[title+'_unit'] = (unit)
    return x
```

```python
def cb_curve(y_true, y_pred):
    xaxis= np.linspace(0, 1, 100)
    #result = np.zeros(xaxis.shape)
    result = { "cb2": np.zeros(xaxis.shape),  "cb": np.zeros(xaxis.shape)}
    for i,x in enumerate(xaxis):
        y_pred2 = y_pred.copy()
        y_pred2[(y_pred2 >= x)] = 1
        y_pred2[(y_pred2 < x)] = 0
        #cb = np.sum(confusion_matrix(y_test, y_pred) * np.array([[0, -1],[ 0, 10]]))
        result["cb"][i] = np.sum(confusion_matrix(y_test, y_pred2) * np.array([[0, -1],[ 0, 9]])) / len(y_true) 
        #result["cb2"][i] = np.sum(confusion_matrix(y_test, y_pred2) * np.array([[0, -1],[ 0, 29]])) / len(y_true)
    
    return (xaxis, result)
```

```python
y_hat =np.array(list(list(zip(*clf.predict_proba(X_test)))[1]))
x, result = cb_curve(y_test, y_hat)
plt.plot(x, result['cb'], label='assumed average 10x')
plt.xlabel('prediction Exit Threshold')
plt.ylabel('expected return')
plt.title("Cost Benefit Analysis")
plt.legend()
plt.savefig("cost_benefit.png")
```

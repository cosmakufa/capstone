```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_csv('../data/cb_objects.csv',parse_dates=['founded_at'])
```

```python
removed = [
 'entity_type',
 'entity_id',
 'parent_id',
 'name',
 
 'permalink',



 'closed_at',

 'homepage_url',
 'twitter_username',
 'logo_url',
 'logo_width',
 'logo_height',
 'short_description',
 'description',
 

 

 'first_investment_at',
 'last_investment_at',
 'investment_rounds',
 'invested_companies',
 'first_funding_at',
 'last_funding_at',
 'funding_rounds',
 'funding_total_usd',
 'first_milestone_at',
 'last_milestone_at',
 'milestones',
 'relationships',
 'created_by',
 'created_at',
 'updated_at']
keep = [x for x in list(df.columns) if x not in removed]
df = df[keep]
df.head().T
```

```python
df.head().T
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
new_df = df[['founded_at', 'category_code', 'id']]
new_df['years_since'] = new_df['founded_at'].apply(lambda x: (pd.Timestamp('01-01-2014') - x).days)
index = list(new_df['category_code'].value_counts(1)[:20].index)
new_df[index] = pd.get_dummies(df['category_code'])[index]
new_df.drop(columns=['category_code','founded_at'], inplace=True)
```

```python
df['tag_list'] = df['tag_list'].apply(lambda x: x.split(", ") if not pd.isna(x) else [])
```

```python
mlb = MultiLabelBinarizer()

result =pd.DataFrame(mlb.fit_transform(df['tag_list']),columns=mlb.classes_, index=df.index)
```

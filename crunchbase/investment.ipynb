```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_investments = pd.read_csv('data/cb_investments.csv', parse_dates=['created_at', 'updated_at'])
```

```python
df_investments.head()
```

```python
df_investments.groupby('funded_object_id').count().sort_values("id", ascending=False).hist('id', bins=100, figsize=(20,10));
```

interesting things to note from this graph is that there is a exponential drop for the number of people funding for startups. keep in mind some companies can repeat investments in different rounds.
